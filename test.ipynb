```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# Replace 'path/to/driver' with the actual path to the driver executable
driver = webdriver.Chrome()
```

```python
driver.get('https://partners.dev.rappi.com/')
```

```python

# Find the element by class name
element = driver.find_elements(By.TAG_NAME, 'body')[0]

# Get the HTML content of the element
html_content = element.get_attribute('innerHTML')


html_content
```

```python
from bs4 import BeautifulSoup
import re

def remove_useful_tags(html_content):
    # Unwrap tags that convey useful information to the user (e.g., <a>, <strong>, <em>, etc.)
    soup = BeautifulSoup(html_content, 'html.parser')
    for tag in soup.find_all(['a', 'strong', 'em', 'b', 'i', 'u']):
        tag.unwrap()  # Removes the tag but keeps its contents

    # Remove tags that do not convey useful/meaningful information to the user
    for tag in soup.find_all(['script', 'noscript', 'style']):
        tag.extract()
    
    # Remove unwanted attributes from the remaining tags
    for tag in soup.find_all(True):
        tag.attrs = {attr: value for attr, value in tag.attrs.items() if attr in ['href', 'src', 'class', 'id']}

    # Remove any remaining HTML tags
    # clean_text = soup.get_text()

    # Remove multiple spaces and newlines


    # clean_text = re.sub(r'\s+', ' ', soup).strip()

    return soup.prettify()

cleaned_text = remove_useful_tags(html_content)
print(cleaned_text)
```
